# Using nGibbs in PetThermoTools

nGibbs is a neural network machine learning model (https://github.com/benjthyer/nGibbs) that is trained on rhyolite-MELTS 1.0.2 and rhyolite-MELTS 1.2.0. 

## Why use an emulator instead of "the real deal"? 

nGibbs is 100s to 1000s of times as fast as MELTS for large jobs. Some problems are much easier to solve when the calculations are fast. For instance, Let's say you have a cumulate rock and you want to identify the composition of its parent magma and the conditions of its crystallization. There's a lot of variables that you may need to explore: Pressure, temperature, oxygen fugacity, water content, SiO2, CaO, Al2O3, CO2, etc etc... You could use a Monte Carlo Markov chain with all of these degrees of freedom, but you would need a very fast model for this to be feasible! This is where nGibbs comes in. Surrogate models like nGibbs are also used for making geodynamic simulations more realistic. A fast model like nGibbs is also useful for broad searches of compositions and conditions in early phases of research in preparation for full thermodynamic calculations.

Although not implemented in PetThermoTools, nGibbs also emulates HeFESTo thermodynamic modelling for subsolidus phase equilibria of the mantle (https://kgblab.epss.ucla.edu/project/hefesto/). Releases coming soon will support pMELTS as well! 

This release is a bit of a preview, so please send Ben comments, notice of bugs, suggestions, collaborations, and questions! If nGibbs isn't quite there yet for your specific use, we may be able to train a new model for it. (bthyer@caltech.edu)

One of the coolest things about nGibbs is that it can be used in PetThermoTools very similarly to the other MELTS models, so it should be easy to integrate into your workflows. 

For instance, instead of calling 

```
equil = ptt.equilibrate_multi(Model = "MELTSv1.2.0",
                            bulk = LL8,
                            P_bar = Sat_PT['P_bar'].values,
                            T_C = Sat_PT['T_Liq_C'].values)
```

Just put an "n" in front of the Model name: "nMELTSv1.2.0"! 
And optionally add "NoProp" afterwards if you can skip the calculations of the thermodynamic properties (this is useful if you are only interested in phase equilibria: It saves a lot of time.)

This runs the same calculation on the nGibbs emulator of rhyolite-MELTS 1.2.0!

```
equil = ptt.equilibrate_multi(Model = "nMELTSv1.2.0", # or Model = "nMELTSv1.2.0NoProp"
                            bulk = LL8,
                            P_bar = Sat_PT['P_bar'].values,
                            T_C = Sat_PT['T_Liq_C'].values)
```

### 1.2 - Melt inclusion saturation pressures

PetThermoTools provides you with 2 ways to calculate melt inclusion saturation pressures. Notably, in both methods you need to provide the CO$_2$ contents in wt% (combining the CO$_2$ contribute from both the melt phase and bubble if present).

1. Simple saturation pressure calculations - we can use the rhyolite-MELTS v1.2.0 model (Ghiorso and Gualda, 2015) to calculate saturation pressure assuming a known temperature (`T_fixed_C` - usually determined using an empirical, liquid-based thermometer). All solid phases are turned off for these calculations, allowing the system to only stabilze liquid and fluid phases. The system will then iterate pressure to find the point of volatile saturation for each composition provided. This should be relatively fast. **Run the cell below and examine the structure of the outputs.** This method is essentially the same as that used in VESIcal when the magmasat model is selected (Iacovino et al. 2021).

````python
Sat_P = ptt.saturation_pressure(Model = "MELTSv1.2.0",
                                bulk = LL8,
                                P_init_bar = 1000,
                                T_fixed_C = LL8['Temp'])

Sat_P.head()
````

This works well when you simply want the saturation pressure. But what if you intend to run further calculations using these compositions? In that case, its more useful to use PetThermoTools to solve for both pressure and temperature.

2. Coupled saturation pressure and liquidus calculations - one of the advanatages of PetThermoTools compared to VESIcal is the ability to deal with solid-melt-fluid systems, rather than a 2-component melt-fluid system. This means we can provide an initial temperature (`T_init_C`) rather than a fixed temperature and the system with iteratively solve for both T & P, searching for the highest pressure at which volatie saturation occurs at the liquidus temperature. **Complete the cell below to calculate the saturation pressure and liquidus temperature of these melt inclusion compositions. Use the following cells below to plot the two different pressure estimates and two different temperature estimates against each other.**

````python
Sat_PT = ptt.saturation_pressure(Model = "MELTSv1.2.0",
                                bulk = LL8,
                                P_init_bar = 1000,
                                T_init_C = LL8['Temp'] + 50)
````

````python
f, a = plt.subplots(1,2, figsize = (8, 3.5))
a[0].plot(Sat_P['P_bar'], Sat_PT['P_bar'], 'ok')
a[0].set_ylabel('P (bar); P & T solved')
a[0].set_xlabel('P (bar); P only')

a[1].plot(Sat_P['T_Liq_C'], Sat_PT['T_Liq_C'], 'ok')
a[1].set_ylabel('T ($^o$C); P & T solved')
a[1].set_xlabel('T ($^o$C); P only')

f.tight_layout()
````

### 1.3 - Equilibration calculations and output structure

What we can do, now that we have a composition, pressure, and temperature for each inclusion, is run a set of simple equilibration calculations. Given that we searched for the P & T conditions of volatile saturation at the liquidus above, what these calculations will hopefully show is that the system is almost entirely melt/liquid at these conditions with a small about of fluid and a single solid phase. 

**Run the `equilibrate_multi` function below and examine the outputs. Specifically have a look at the `mass_g` DataFrame. What is the liquidus phase for these compositions? Is this expected?**

*nGibbs interjection: The speed of equilibrate_multi is massively improved by nGibbs, because the full thermodynamic calculations here require individual instances of MELTS to be initialized, and each in turn requires a costly liquidus-find / subsolidus initial guess. nGibbs evaluates everything in perfect parallel isolation, even below the liquidus. Therefore, there is no additional expense incurred by running multiple compositions, nor any relative speedup for calculations at a constant composition.*

*Note also that nGibbs is not able to suppress phases at this time, and thus is not supported for the above ptt.saturation_pressure() function*

````python
start_time = time()
equil = ptt.equilibrate_multi(Model = "MELTSv1.2.0",
                            bulk = LL8,
                            P_bar = Sat_PT['P_bar'].values,
                            T_C = Sat_PT['T_Liq_C'].values)
end_time_PTT = time()
print(f"PTT equilibration time: {end_time_PTT - start_time:.2f} seconds")

try:
    n_equil = ptt.equilibrate_multi(Model = "nMELTSv1.2.0NoProp", # No properties needed for this analysis
                                bulk = LL8,
                                P_bar = Sat_PT['P_bar'].values,
                                T_C = Sat_PT['T_Liq_C'].values)

    print(f"nGibbs equilibration time: {time() - end_time_PTT:.2f} seconds")
except:
    print("Install nGibbs above and restart to run nGibbs through PTT!")
````

### 2.2 - Isobaric crystallization

Likely the most common function used in PetThermoTools is the `isobaric_crystallisation` function (note that it is an 's' not a 'z' because I'm British). So let's learn how to use that!

There are many options in the `isobaric_crystallization` function. We need to provide information about pressure (`P_bar`), the starting composition (`bulk`) and the thermodynamic model (`Model`). In addition, we need to specify temperature constraints, information about the oxygen fugacity of the system, and whether we wish to perform fractional or equilibrium crystallization. There are many other options, which you can learn about through our documentation.

**Using the code below perform 4 equilibrium crystallization calculations at different pressures. Try running similar calculations using fractional crystallization instead.**

````python
start_time = time()

isobars = [250,500,1000,2000]

isobaric = ptt.isobaric_crystallisation(Model = "MELTSv1.0.2",
                                        bulk = Kil_starting,
                                        P_bar = isobars,
                                        find_liquidus = True,
                                        T_end_C = 1050,
                                        dt_C = 2,
                                        label = "P_bar")

end_time_PTT = time()
print(f"MELTS 1.0.2 isobaric crystallisation time: {end_time_PTT - start_time:.2f} seconds")

try:
    n_isobaric = ptt.isobaric_crystallisation(Model = "nMELTSv1.0.2NoProp", # don't type NoProps (with an s, easy mistake) or it will silently route to full MELTSv1.0.2!
                                            bulk = Kil_starting,
                                            P_bar = isobars,
                                            find_liquidus = True,
                                            T_end_C = 1050,
                                            dt_C = 2,
                                            label = "P_bar")
    print(f"nMELTS 1.0.2 isobaric crystallisation time: {time() - end_time_PTT:.2f} seconds")
except:
    pass
````

Use the cell below to examine the outputs (hint use `isobaric.keys()` to see how the different calculations are stored in the output dictionary). 

````python
# Let's plot the phase abundances calculated from MELTS and nGibbs!(This will error if you don't have nGibbs installed))
for P in isobars:
    fig, (ax_a, ax_b) = plt.subplots(2,1, figsize=(16, 5),
                                    sharex=True, constrained_layout=True)

    ptt.Plotting._draw_stack(ax_a, isobaric[f'P = {P:.1f} bars']['Conditions']['T_C'], isobaric[f'P = {P:.1f} bars']['mass_g']/100,
                    isobaric[f'P = {P:.1f} bars']['mass_g'].columns, colors=ptt.Plotting.melts_colors(), title=f'PetThermoTools MELTS 1.0.2 Kilauea at P = {int(P)} bars',
                    x_name= '')
    ptt.Plotting._draw_stack(ax_b, n_isobaric[f'P = {P:.1f} bars']['Conditions']['T_C'], n_isobaric[f'P = {P:.1f} bars']['mass_g']/100,
                    n_isobaric[f'P = {P:.1f} bars']['mass_g'].columns, colors=ptt.Plotting.melts_colors(), title=f'nGibbs MELTS 1.0.2 Kilauea at P = {int(P)} bars',
                    x_name= '')
    plt.show()
````

The liquidus is a bit high in the nGibbs version. More prominently, nGibbs seems reluctant to predict small amounts of phases, and therefore phases don't gradually pinch in and out at saturation like they should, instead nGibbs produces 'stairsteps' of phase accumulation at saturation, similar to what would be expected in many real world scenarios due supercooling past the saturation point due to the kinetic barrier to crystallisation/bubble nucleation. But in equilibrium thermodynamics, it is an error. 

**Use the `ptt.harker()` function to plot the results. We can also plot the matrix glass data from Wieser et al. (2021) using this function. Are these calculations doing a good job? If you're waiting, try different H2O or fO2 settings, and/or try different thermodynamic models.**

## Part 3 - Phase Diagrams

One interesting observation from the Hawaiian islands is that high-MgO orthopyroxene crystals are sometimes observed in lavas from the Loa trend (Mauna Loa, Hualalai, etc.) but these are rare or completely absent in samples from the Kea trend (Kilauea, Mauna Kea, Haleakala, etc.). The question becomes, is this something we can predict through thermodynamics, given that there are subtle differences in the major element composition of primary melts from each trend (see Fig. 8a in Wieser et al. 2025 for more details; https://doi.org/10.1007/s00445-025-01869-2).

**Using the Kilauea starting composition we defined earlier calculate and plot the phase diagram for this composition at pressures between 250 and 5000 bar and temperatures between 1100 and 1400 Celsius (hint, use `np.linspace`)**

````python
begin_time = time()
PD_Kil = ptt.phaseDiagram_calc(Model = "MELTSv1.2.0",
                                bulk = Kil_starting,
                                T_C = np.linspace(1100,1400,20),
                                P_bar = np.linspace(250,5000,20),
                                refine = 2,
                                i_max = 3)

print(f"Phase diagram calculation time: {time() - begin_time:.2f} seconds")
````

````python
ptt.plot_phaseDiagram(Combined = PD_Kil)
````

````python
begin_time = time()
nPD_Kil = ptt.phaseDiagram_calc(Model = "nMELTSv1.2.0",
                                bulk = Kil_starting,
                                T_C = np.linspace(1100,1400,20),
                                P_bar = np.linspace(250,5000,20),
                                refine = 2,
                                i_max = 10
                                )

print(f"nGibbs Phase diagram calculation time: {time() - begin_time:.2f} seconds")

````

````python
ptt.plot_phaseDiagram(Combined = nPD_Kil)
````

````python
ptt.plot_phaseDiagram_multi({'nGibbs': nPD_Kil, 'rhyolite-MELTS v1.0.2': PD_Kil})
````

Finally, we can use the following code to extract information from the phase diagrams. In this case let's look directly at the mass of orthopyroxene in each calculation:

````python
X_Kil, Y_Kil, Z_Kil = ptt.make_grid_variables_from_phase_diagram(
    Results = PD_Kil,
    x_col = "T_C",
    y_col = "P_bar",
)

X_ML, Y_ML, Z_ML = ptt.make_grid_variables_from_phase_diagram(
    Results = PD_ML,
    x_col = "T_C",
    y_col = "P_bar",
)

X_nKil, Y_nKil, Z_nKil = ptt.make_grid_variables_from_phase_diagram(
    Results = nPD_Kil.sort_values(['T_C', 'P_bar'], ascending=[True, True]),
    x_col = "T_C",
    y_col = "P_bar",
)

X_nML, Y_nML, Z_nML = ptt.make_grid_variables_from_phase_diagram(
    Results = nPD_ML.sort_values(['T_C', 'P_bar'], ascending=[True, True]),
    x_col = "T_C",
    y_col = "P_bar",
)
````
