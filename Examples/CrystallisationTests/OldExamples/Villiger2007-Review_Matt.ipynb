```python
Exp = pd.read_excel('Villiger2007.xlsx')
Elements = ['SiO2_Liq', 'TiO2_Liq', 'Al2O3_Liq', 'FeOt_Liq', 'MgO_Liq',
            'CaO_Liq', 'Na2O_Liq', 'K2O_Liq']
```

```python
bulk = {'SiO2_Liq': Exp['SiO2_Liq'].loc[0], 
        'TiO2_Liq': Exp['TiO2_Liq'].loc[0], 
        'Al2O3_Liq': Exp['Al2O3_Liq'].loc[0], 
        'FeOt_Liq': Exp['FeOt_Liq'].loc[0], 
        'MnO_Liq': Exp['MnO_Liq'].loc[0], 
        'MgO_Liq': Exp['MgO_Liq'].loc[0], 
        'CaO_Liq': Exp['CaO_Liq'].loc[0], 
        'Na2O_Liq': Exp['Na2O_Liq'].loc[0], 
        'K2O_Liq': Exp['K2O_Liq'].loc[0],
        'Cr2O3_Liq': 0.01,
        'P2O5_Liq': 0.0,
        'H2O_Liq': 0.2, 
        'CO2_Liq': 0.0,
       'Fe3Fet_Liq': 0.062}

bulk
```

```python
Exp[Elements] = 100*Exp[Elements].values/np.nansum(Exp[Elements], axis = 1).reshape(8,1)
```

```python
P_bar = np.array([1000,4000,7000,10000])
```

```python
Results_Holland = M.multi_path(Model = "Holland", comp = bulk,
                       T_end_C = 1050, dt_C = 2, 
                       P_bar = P_bar, find_liquidus = True,
                       H2O_Liq = 0.2, Frac_solid = True,
                              timeout = 600, Fe3Fet_Liq = np.array([0.064, 0.064, 0.064, 0.064]))

Results_MELTSv120 = M.multi_path(Model = "MELTSv1.2.0", comp = bulk,
                       T_end_C = 1000, dt_C = 2, 
                       P_bar = P_bar, find_liquidus = True,
                       H2O_Liq = 0.2,
                       Frac_solid = True, Frac_fluid = True,
                      fO2_buffer = "FMQ", fO2_offset = -2,
                              timeout = 600)

Results_MELTSv102 = M.multi_path(Model = "MELTSv1.0.2", comp = bulk,
                       T_end_C = 1000, dt_C = 2, 
                       P_bar = P_bar, find_liquidus = True,
                       H2O_Liq = 0.2,
                       Frac_solid = True, Frac_fluid = True,
                      fO2_buffer = "FMQ", fO2_offset = -2,
                              timeout = 300)

Results_pMELTS = M.multi_path(Model = "pMELTS", comp = bulk,
                       T_end_C = 1000, dt_C = 2, 
                       P_bar = P_bar, find_liquidus = True,
                       H2O_Liq = 0.2,
                       Frac_solid = True, Frac_fluid = True,
                      fO2_buffer = "FMQ", fO2_offset = -2,
                              timeout = 600)
```

```python
S = {'index = 0': 1, 
     'index = 1': 4,
     'index = 2': 7,
     'index = 3': 10}

for s in S:
    Columns_bad = ['SiO2', 'TiO2', 'Al2O3', 'FeOt', 'MnO', 'MgO', 'CaO', 'Na2O', 'K2O', 'P2O5', 'H2O', 'Cr2O3', 'Fe3Fet']
    Columns_ideal = ['SiO2_Liq', 'TiO2_Liq', 'Al2O3_Liq', 'FeOt_Liq', 'MnO_Liq', 'MgO_Liq', 'CaO_Liq', 'Na2O_Liq', 'K2O_Liq', 'P2O5_Liq', 'H2O_Liq', 'Cr2O3_Liq', 'Fe3Fet_Liq']

    Comp_start = Results_Holland[s]['liq'].copy()
    for el in Comp_start:
        if el in Columns_bad:
            Results_Holland[s]['liq'] = Results_Holland[s]['liq'].rename(columns = {el:el + '_Liq'})

    for el in Columns_ideal:
        if el not in list(Results_Holland['index = 0']['liq'].keys()):
            Results_Holland[s]['liq'][el] = np.zeros(len(Results_Holland[s]['liq'].iloc[:,0]))

    Results_Holland[s]['liq']['FeOt_Liq'] = Results_Holland[s]['liq']['FeO'] 
    import Thermobar as pt
    A = pt.convert_fe_partition_to_fo2(liq_comps = Results_Holland[s]['liq'],
                                   T_K = Results_Holland[s]['Conditions']['T_C'] + 273.15,
                                   P_kbar = S[s],
                                   model = "Kress1991",
                                   renorm = False)
    
    Results_Holland[s]['liq']['deltaQFM'] = A['DeltaQFM_Frost1991']
```

```python
f, a = plt.subplots(1,1)
for i in Results_Holland:
    a.plot(Results_Holland[i]['liq']['MgO_Liq'],
           Results_Holland[i]['liq']['deltaQFM'], '-', label = i)
    
a.legend()
```
